# ELU
指数线性单元: [Fast and Accurate Deep Network Learning by Exponential Linear Units (ELUs)](https://arxiv.org/abs/1511.07289).  

定义：  
torch.nn.ELU(alpha=1.0, inplace=False)


$$\text{ELU}(x) = \begin{cases}
        x, & \text{ if } x > 0\\
        \alpha * (\exp(x) - 1), & \text{ if } x \leq 0
        \end{cases}$$


参数:  
alpha (float) – the α value for the ELU formulation. Default: 1.0
ELU公式的α值

inplace (bool) – can optionally do the operation in-place. Default: False
就地操作

```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
import plotly.express as px

input = torch.arange(start=-10,end=10,step=0.1, dtype=torch.float)

m = nn.ELU(alpha=1.0, inplace=False)

output = m(input)
plt.plot(input,output)
plt.show()

# fig = px.scatter(x=input, y=output)
# fig.show()
```
