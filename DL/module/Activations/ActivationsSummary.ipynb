# ELU
指数线性单元，参考：[Fast and Accurate Deep Network Learning by Exponential Linear Units (ELUs)](https://arxiv.org/abs/1511.07289).  

公式：  
$$\text{ELU}(x) = \begin{cases}
        x, & \text{ if } x > 0\\
        \alpha * (\exp(x) - 1), & \text{ if } x \leq 0
\end{cases}$$

定义：  
torch.nn.ELU(alpha=1.0, inplace=False)

参数:  
- alpha (float) – the α value for the ELU formulation. Default: 1.0.  ELU公式的α值

- inplace (bool) – can optionally do the operation in-place. Default: False.    就地操作

```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt

input = torch.arange(start=-10,end=10,step=0.4, dtype=torch.float)

m = nn.ELU(alpha=1.0, inplace=False)

output = m(input)
plt.scatter(input,output,marker=".")
plt.show()
```

# Hardshrink

公式：  
> $$\text{HardShrink}(x) =
        \begin{cases}
        x, & \text{ if } x > \lambda \\
        x, & \text{ if } x < -\lambda \\
        0, & \text{ otherwise }
\end{cases}$$

定义：  
> torch.nn.Hardshrink(lambd=0.5)

参数:  
> - lambd (float) – the λ value for the Hardshrink formulation. Default: 0.5  
> 硬收缩公式的λ值。默认值:0.5

```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt

input = torch.arange(start=-2,end=2,step=0.1, dtype=torch.float)

m = nn.Hardshrink(lambd=0.5)

output = m(input)
plt.scatter(input,output,marker=".")
plt.show()
```

# Hardsigmoid

公式：  
> $$\text{Hardsigmoid}(x) = \begin{cases}
            0 & \text{if~} x \le -3, \\
            1 & \text{if~} x \ge +3, \\
            x / 6 + 1 / 2 & \text{otherwise}
\end{cases}$$

定义：  
> torch.nn.Hardsigmoid(inplace=False)

参数:  
> - inplace (bool) – can optionally do the operation in-place. Default: False  
> 硬收缩公式的λ值。默认值:0.5

```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt

input = torch.arange(start=-10,end=10,step=0.4, dtype=torch.float)

m = nn.Hardsigmoid()

output = m(input)
plt.scatter(input,output,marker=".")
plt.show()
```

# Hardtanh

公式：  
> $$\text{HardTanh}(x) = \begin{cases}
            \text{max\_val} & \text{ if } x > \text{ max\_val } \\
            \text{min\_val} & \text{ if } x < \text{ min\_val } \\
            x & \text{ otherwise } \\
\end{cases}$$

定义：  
> torch.nn.Hardtanh(min_val=- 1.0, max_val=1.0, inplace=False, min_value=None, max_value=None)

参数:  
> - min_val (float) – minimum value of the linear region range. Default: -1  
> 线性区域范围的最小值
> 
> - max_val (float) – maximum value of the linear region range. Default: 1  
> 线性区域范围的最大值
> 
> - inplace (bool) – can optionally do the operation in-place. Default: False  
> 就地操作

```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt

input = torch.arange(start=-10,end=10,step=0.3, dtype=torch.float)

m = nn.Hardtanh(min_val=- 1.0, max_val=1.0, inplace=False, min_value=-5, max_value=5)

output = m(input)
plt.scatter(input,output,marker=".")
plt.show()
```

# Hardswish

可参考[MobileNetV3](https://arxiv.org/abs/1905.02244).

公式：  
> $$\text{Hardswish}(x) = \begin{cases}
            0 & \text{if~} x \le -3, \\
            x & \text{if~} x \ge +3, \\
            x \cdot (x + 3) /6 & \text{otherwise}
\end{cases}$$

定义：  
> torch.nn.Hardswish(inplace=False)

参数:  
> - inplace (bool) – can optionally do the operation in-place. Default: False  
> 就地操作

```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt

input = torch.arange(start=-10,end=10,step=0.3, dtype=torch.float)

m = nn.Hardswish()

output = m(input)
plt.scatter(input,output,marker=".")
plt.show()
```

# LeakyReLU

公式：  
> $$\text{LeakyReLU}(x) =
        \begin{cases}
        x, & \text{ if } x \geq 0 \\
        \text{negative\_slope} \times x, & \text{ otherwise }
\end{cases}$$

定义：  
> torch.nn.LeakyReLU(negative_slope=0.01, inplace=False)  

参数:  
> - negative_slope (float) – Controls the angle of the negative slope. Default: 1e-2  
> 控制负斜率的角度  
> 
> - inplace (bool) – can optionally do the operation in-place. Default: False  
> 就地操作

```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt

input = torch.arange(start=-10,end=10,step=0.3, dtype=torch.float)

m = nn.Hardswish()

output = m(input)
plt.scatter(input,output,marker=".")
plt.show()
```
