# ELU
指数线性单元，参考：[Fast and Accurate Deep Network Learning by Exponential Linear Units (ELUs)](https://arxiv.org/abs/1511.07289).  

公式：  
$$\text{ELU}(x) = \begin{cases}
        x, & \text{ if } x > 0\\
        \alpha * (\exp(x) - 1), & \text{ if } x \leq 0
\end{cases}$$

定义：  
torch.nn.ELU(alpha=1.0, inplace=False)

参数:  
- alpha (float) – the α value for the ELU formulation. Default: 1.0.  ELU公式的α值

- inplace (bool) – can optionally do the operation in-place. Default: False.    就地操作

```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt

input = torch.arange(start=-10,end=10,step=0.4, dtype=torch.float)

m = nn.ELU(alpha=1.0, inplace=False)

output = m(input)
plt.scatter(input,output,marker=".")
plt.show()
```

# Hardshrink

公式：  
$$\text{HardShrink}(x) =
        \begin{cases}
        x, & \text{ if } x > \lambda \\
        x, & \text{ if } x < -\lambda \\
        0, & \text{ otherwise }
\end{cases}$$

定义：  
torch.nn.Hardshrink(lambd=0.5)

参数:  
- lambd (float) – the λ value for the Hardshrink formulation. Default: 0.5.  硬收缩公式的λ值。默认值:0.5

```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt

input = torch.arange(start=-2,end=2,step=0.1, dtype=torch.float)

m = nn.Hardshrink(lambd=0.5)

output = m(input)
plt.scatter(input,output,marker=".")
plt.show()
```

# Hardtanh

公式：  
$$\text{HardTanh}(x) = \begin{cases}
            \text{max\_val} & \text{ if } x > \text{ max\_val } \\
            \text{min\_val} & \text{ if } x < \text{ min\_val } \\
            x & \text{ otherwise } \\
\end{cases}$$

定义：  
torch.nn.Hardtanh(min_val=- 1.0, max_val=1.0, inplace=False, min_value=None, max_value=None)

参数:  
- min_val (float) – minimum value of the linear region range. Default: -1.  线性区域范围的最小值
- max_val (float) – maximum value of the linear region range. Default: 1.  线性区域范围的最大值
- inplace (bool) – can optionally do the operation in-place. Default: False.  就地操作

```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt

input = torch.arange(start=-10,end=10,step=0.3, dtype=torch.float)

m = nn.Hardtanh(min_val=- 1.0, max_val=1.0, inplace=False, min_value=-5, max_value=5)

output = m(input)
plt.scatter(input,output,marker=".")
plt.show()
```

# PReLU
该激活函数可以自适应地学习矫正线性单元的参数，并且能够在增加可忽略的额外计算成本下提高准确率。

公式：  
$$\text{PReLU}(x) =
        \begin{cases}
        x, & \text{ if } x \geq 0 \\
        ax, & \text{ otherwise }
\end{cases}$$

定义：  
torch.nn.PReLU(num_parameters=1, init=0.25, device=None, dtype=None)

参数:  
- num_parameters (int) – number of a to learn. Although it takes an int as input, there is only two values are legitimate: 1, or the number of channels at input. Default: 1.   可以接受的参数量，允许不同通道的激活函数不一样。可以选择的值有1或通道数。

- init (float) – the initial value of a. Default: 0.25.  a的初始值  

```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt

input = torch.arange(start=-10,end=10,step=0.3, dtype=torch.float)

m = nn.PReLU(num_parameters=1, init=0.25)

output = m(input)
plt.scatter(input,output.detach(),marker=".")
plt.show()
```

# SELU
scaled ELU，对ELU做了修正，参考：[Self-Normalizing Neural Networks](https://arxiv.org/abs/1706.02515)

公式：  
$$\text{SELU}(x) = \text{scale} * (\max(0,x) + \min(0, \alpha * (\exp(x) - 1))).$$
其中，$\alpha = 1.6732632423543772848170429916717$，$scale = 1.0507009873554804934193349852946$.

定义：  
torch.nn.SELU(inplace=False)

参数:  
- inplace: can optionally do the operation in-place. Default: ``False``.  就地操作

```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt

input = torch.arange(start=-10,end=10,step=0.3, dtype=torch.float)

m = nn.SELU()

output = m(input)
plt.scatter(input,output,marker=".")
plt.show()
```

# CELU
Continuously Differentiable ELU，让ELU连续可微分，参考：[Continuously Differentiable Exponential Linear Units](https://arxiv.org/abs/1704.07483)

公式：  
$$\text{CELU}(x) = \max(0,x) + \min(0, \alpha * (\exp(x/\alpha) - 1))$$

定义：  
torch.nn.CELU(alpha=1.0, inplace=False)

参数:  
- alpha: the $\alpha$ value for the CELU formulation. Default: 1.0.  CELU公式的$\alpha$值
- inplace: can optionally do the operation in-place. Default: ``False``.  就地操作

```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt

input = torch.arange(start=-10,end=10,step=0.3, dtype=torch.float)

m = nn.CELU(alpha=1.0)

output = m(input)
plt.scatter(input,output,marker=".")
plt.show()
```

# GELU

公式：  
$$\text{GELU}(x) = x * \Phi(x)$$
其中，$\Phi(x)$ 是高斯分布的累计分布函数。

定义：  
torch.nn.GELU(approximate='none')

参数:  
- approximate (str, optional) – the gelu approximation algorithm to use: 'none' | 'tanh'. Default: 'none'.  使用的GELU近似算法:'none' | 'tanh'。默认值:“没有”。


```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt

input = torch.arange(start=-10,end=5,step=0.3, dtype=torch.float)

m = nn.GELU(approximate='none')

output = m(input)
plt.scatter(input,output,marker=".")
plt.show()
```

# Softplus

参数:  
- beta: the $\beta$ value for the Softplus formulation. Default: 1.  Softplus公式的$\beta$值。

- threshold: values above this revert to a linear function. Default: 20.  高于此值的值恢复为线性函数。默认值:20

```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt

input = torch.arange(start=-10,end=10,step=0.3, dtype=torch.float)

m = nn.Softplus(beta=1, threshold=5)

output = m(input)
plt.scatter(input,output,marker=".")
plt.show()
```

# Threshold

公式：  
$$y =
        \begin{cases}
        x, &\text{ if } x > \text{threshold} \\
        \text{value}, &\text{ otherwise }
\end{cases}$$

定义：  
torch.nn.Threshold(threshold, value, inplace=False)

参数：  
- threshold: The value to threshold at.  阈值
- value: The value to replace with.  要替换的值
- inplace: can optionally do the operation in-place. Default: ``False``.  就地操作

```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt

input = torch.arange(start=-10,end=10,step=0.3, dtype=torch.float)

m = nn.Threshold(threshold=3, value=2, inplace=False)

output = m(input)
plt.scatter(input,output,marker=".")
plt.show()
```
